```python
#Test Initialisierung und show()
import ListKeeper

a = ListKeeper.ListKeeper()

help(a.show)

assert a.show() == ['example']
```

```python
#Test add
help(a.add)
a.add('test1', [2,3,2,3,2,3])
assert a.show() == ['example', 'test1']
```

```python
#Test sort
help(a.sort)
assert a.sort('test1') == [2, 2, 2, 3, 3, 3]
```

```python
#Test append
help(a.append)
a.append('test1', [1,6])
assert a.sort('test1') == [1, 2, 2, 2, 3, 3, 3, 6]
```

```python
#Test delete
help(a.delete)
a.delete('test1')
assert a.show() == ['example']
```
