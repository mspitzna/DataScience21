```python
#import an read big-mart data
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#Read data into pandas table - details on pandas in block 3
data = pd.read_csv("../../DATA/company_sales_data.csv") 
```

## Exercise 1
Read Total profit of all months and show it using a line plot
Total profit data provided for each month. Generated line plot must include the following properties: 

* X label name = Month Number
* Y label name = Total profit

### The result should look something like this:
<img src="IMG/1.png" width=400>

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(data.month_number, data.total_profit, 'teal')

axes.set_ylim([100000, 500000])

axes.set_xlabel('Month Number')
axes.set_ylabel('Total profit')

axes.set_title('Company profit per month');
```

## Exercise 2
Get Total profit of all months and show line plot with the following Style properties
Generated line plot must include following Style properties:

* Line Style dotted and Line color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3


### The result should look something like this:
<img src="IMG/2.png" width=400>

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # left, bottom, width, height (range 0 to 1)

axes.plot(data.month_number, data.total_profit, 
          label='Profit data of last year', 
          lw = 3, 
          ls='--', 
          marker='o', 
          color='red', 
          markerfacecolor="blue", 
          markersize=6, 
          markeredgecolor="red")

axes.set_ylim([100000, 500000])

axes.set_xticks(data.month_number)
axes.set_xlabel('Month Number')
axes.set_ylabel('Sold units number')

axes.set_title('Company Sales data of last year');

axes.legend(loc=4)
```

## Exercise 3
Read all product sales data and show it  using a multiline plot
Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).


### The result should look something like this:
<img src="IMG/3.png" width=400>

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])


for product in ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']:
    axes.plot(data.month_number, data[product], label=product, lw=3, marker='o')

    
axes.set_xticks(data.month_number)
axes.set_ylim([1000, 18000])
axes.legend()

axes.set_xlabel('Month Number')
axes.set_ylabel('Sales units in number')
axes.set_title('Sales data')
```

## Exercise 4
Read toothpaste sales data of each month and show it using a scatter plot
Also, add a grid in the plot.


### The result should look something like this:
<img src="IMG/4.png" width=400>

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.scatter(data.month_number, data.toothpaste, label='Tooth paste Sales data')

axes.set_xticks(data.month_number)

axes.grid(linestyle='dashed')
axes.legend()
axes.set_xlabel('Month Number')
axes.set_ylabel('Number of units sold')
axes.set_title('Tooth paste Sales data')

```

## Exercise 5
Read face cream and facewash product sales data and show it using the bar chart
Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.


### The result should look something like this:
<img src="IMG/5.png" width=400>

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

width = 0.25 

axes.bar(data.month_number - width / 2, data.facecream, width=width,
         label='Tooth paste Sales data')
axes.bar(data.month_number + width / 2, data.facewash, width=width,
         label='Tooth paste Sales data')

axes.set_xticks(data.month_number)

axes.grid(linestyle='dashed')
axes.legend(loc=2)

axes.set_xlabel('Month Number')
axes.set_ylabel('Sales units inn number')
axes.set_title('Facewash and facecream sales data')
```

## Exercise 6
Calculate total sale data for last year for each product and show it using a Pie chart
Note: In Pie chart display Number of units sold per year for each product in percentage.

***Note:*** use ``data[item].sum()`` to compute sums over columns 


### The result should look something like this:
<img src="IMG/6.png" width=400>

```python
data['toothpaste'].sum()
```

```python
fig = plt.figure(facecolor='white')
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

product_names = ['FaceCream', 'FaceWash', 'ToothPaste', 'Bathing soap', 'Shampoo', 'Moisturizer']
products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
values = list()

for product in products:
    values.append(data[product].sum())

axes.pie(values, labels=product_names, autopct='%1.1f%%')
axes.axis('equal')
axes.legend(loc=4)
axes.set_title('Sales data')

```
