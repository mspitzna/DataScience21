### Step 3. Assign it to a variable called drinks.

```python
drinks=pd.read_csv('../../DATA/drinks.csv')
drinks = drinks.fillna('NA')
```

### Step 4. Which continent drinks the most beer on average?

```python
drinks_grouped = drinks.groupby(drinks['continent'])
drinks_grouped['beer_servings'].mean().sort_values(ascending=False).plot.bar()
```

### Step 5. For each continent print the statistics for wine consumption.

```python
drinks_grouped['wine_servings'].describe()
```

### Step 6. Print the mean alcohol consumption per continent for every column

```python
drinks_grouped.mean()
```

### Step 7. Print the median alcoohol consumption per continent for every column

```python
drinks_grouped.median()
```

### Step 8. Print the mean, min and max values for spirit consumption.
#### This time output a DataFrame

```python
drinks_grouped['spirit_servings'].agg(['mean', 'min', 'max'])
```
